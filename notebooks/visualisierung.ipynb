# Visualisierung mit Matplotlib & Pandas

Starte den interaktiven Modus über das **⚡ Power-Symbol** oben rechts.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sinuswelle plotten
x = np.linspace(0, 2 * np.pi, 300)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, np.sin(x), label='sin(x)', color='steelblue', linewidth=2)
ax.plot(x, np.cos(x), label='cos(x)', color='tomato', linewidth=2)
ax.set_title('Sinus und Kosinus')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Pandas — Daten analysieren

```python
import pandas as pd

# Beispiel-Datensatz
daten = pd.DataFrame({
    'Monat': ['Jan', 'Feb', 'Mär', 'Apr', 'Mai', 'Jun'],
    'Umsatz': [12000, 15000, 13500, 17000, 19500, 21000],
    'Kosten': [9000, 10500, 9800, 11200, 12500, 13000],
})
daten['Gewinn'] = daten['Umsatz'] - daten['Kosten']
print(daten.to_string(index=False))
```

```python
# Balkendiagramm
fig, ax = plt.subplots(figsize=(8, 4))
x_pos = range(len(daten))
width = 0.3

ax.bar([p - width for p in x_pos], daten['Umsatz'], width, label='Umsatz', color='steelblue')
ax.bar(x_pos,                       daten['Kosten'], width, label='Kosten',  color='tomato')
ax.bar([p + width for p in x_pos], daten['Gewinn'], width, label='Gewinn',  color='seagreen')

ax.set_xticks(list(x_pos))
ax.set_xticklabels(daten['Monat'])
ax.set_title('Umsatz, Kosten & Gewinn')
ax.set_ylabel('Euro')
ax.legend()
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## Interaktives Widget

```python
import ipywidgets as widgets
from IPython.display import display

def plot_welle(frequenz=1.0, amplitude=1.0):
    x = np.linspace(0, 4 * np.pi, 500)
    y = amplitude * np.sin(frequenz * x)
    fig, ax = plt.subplots(figsize=(8, 3))
    ax.plot(x, y, color='steelblue', linewidth=2)
    ax.set_ylim(-3, 3)
    ax.set_title(f'Frequenz: {frequenz:.1f} | Amplitude: {amplitude:.1f}')
    ax.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

widgets.interact(
    plot_welle,
    frequenz=widgets.FloatSlider(min=0.5, max=5.0, step=0.5, value=1.0),
    amplitude=widgets.FloatSlider(min=0.5, max=3.0, step=0.5, value=1.0),
)
```
